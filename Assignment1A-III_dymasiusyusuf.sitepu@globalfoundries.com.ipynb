```python
word = [ '       Python for Data Science ','Python         ','python ','   linux internals        ','Linear algebra ', 'Statistics']
```

```python
word
```

```python
clean_word = []
for w in word: 
    c = "".join(w.strip(" "))
    clean_word.append(c)

print(clean_word)
```

```python
# Now this is the clean word. Basically we apply split to get the char inside the string and join back using join()
clean_word
```

# Question 3:

```python
# I dont know why not just using capitalize or title? I meant I use to upper only the first word because the expected output wants me to have Data Science as caps.
clean_word_capitalized = [word[0].upper()+word[1:] for word in clean_word]
clean_word_capitalized
```

# Question 4:

```python
clean_word_capitalized_underscored = ["_".join(word.split(" ")) for word in clean_word_capitalized]
clean_word_capitalized_underscored
```
