```python
word = [ '       Python for Data Science ','Python         ','python ','   linux internals        ','Linear algebra ', 'Statistics']
```

```python
word
```

```python
clean_word = []
for w in word: 
    c = "".join(w.strip(" "))
    clean_word.append(c)

print(clean_word)
```
