# Defining the Age and Salary

```python
Age = [64, 49, 48, 46, 57, 60, 69, 25,23, 32]   
Salary = [20000, 10000, 5000, 25000, 500, 1000, 300, 30000, 2500,20000] 
```

```python
#There is other libraries as well like maths and statistics similar to numpy
import statistics as stat
import numpy as np

Age_mean = stat.mean(Age)
Age_median = stat.median(Age)
Age_mode = stat.mode(Age)

#Well I guess I can use numpy and numpy is the topic of this study. I will use numpy next
print(f"This is stats mean: {Age_mean}")
print(f"This is stats median: {Age_median}")
print(f"This is stats mode: {Age_mode}")

Age_mean = np.mean(Age)
Age_median = np.median(Age)


print(f"This is numpy mean: {Age_mean}")
print(f"This is numpy median: {Age_median}")

```
